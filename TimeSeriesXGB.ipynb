```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import xgboost as xgb
from google.colab import drive
drive.mount('/content/gdrive')
data = pd.read_csv("gdrive/My Drive/tata.csv")

```

```python
close_prices = data['Close'].values.reshape(-1, 1)
volume = data['Total Trade Quantity'].values.reshape(-1, 1)
data_array = np.hstack((close_prices, volume))

# Preprocess data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data_array)
data['Close_scaled'] = data_scaled[:, 0]

# Split data into training and testing sets
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size,:]
test_data = data.iloc[train_size:,:]

# Define features and target variable
features = ['Close_scaled', 'Total Trade Quantity']
target = 'Close'

# Prepare data for XGBoost
train_dmatrix = xgb.DMatrix(train_data[features], label=train_data[target])
test_dmatrix = xgb.DMatrix(test_data[features], label=test_data[target])

```

```python
# Define XGBoost hyperparameters
params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'eta': 0.01,
    'max_depth': 4,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'gamma': 1
}

# Train XGBoost model
model = xgb.train(params, train_dmatrix, num_boost_round=1000, early_stopping_rounds=10, evals=[(test_dmatrix, 'Test')])

# Make predictions on test set
test_data['Predictions'] = model.predict(test_dmatrix)

# Evaluate model performance
rmse = np.sqrt(mean_squared_error(test_data['Close'], test_data['Predictions']))
print('RMSE:', rmse)

# Train XGBoost model
model = xgb.train(params, train_dmatrix, num_boost_round=1000, early_stopping_rounds=10, evals=[(test_dmatrix, 'Test')])

# Make predictions on test set
test_data['Predictions'] = model.predict(test_dmatrix)

# Evaluate model performance
rmse = np.sqrt(mean_squared_error(test_data['Close'], test_data['Predictions']))
print('RMSE:', rmse)
```
